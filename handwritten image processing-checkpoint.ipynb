## what is handwritten  digits image processing or Handwritten digit recongnization?
- Handwritten digits image processing which is also known as "Handwritten digit recongination" refers to the application of various techniques and algorithms to analyze, enhance, and extract information from images containing handwritten digits.

- This field combines principles from computer vision, image processing, and machine learning to perform tasks such as digit segmentation, noise reduction, feature extraction, and classification. 

- Handwritten digit recognition is the process to provide the ability to machines to recognize human handwritten digits.

- It is not an easy task for the machine because handwritten digits are not perfect, vary from person-to-person, and can be made with many different flavors.


# what is MNIST Dataset?
- MNIST (Modified National Institute of Standards and Technology) is a popular dataset commonly used for training and evaluating machine learning models for handwritten digit recognition. 
- It consists of a large collection of grayscale images of handwritten digits, each labeled with the corresponding digit (0-9).

- The MNIST dataset contains 60,000 training examples and 10,000 testing examples. 
- Each image in the dataset has a fixed size of 28x28 pixels, resulting in a total of 784 pixels per image. 
- The pixel values range from 0 to 255, representing different shades of gray.
-  It has been widely used to develop and evaluate various algorithms, including deep learning models, for handwritten digit recognition.

# Preprocessing the image

- Model cannot take the image data directly so we need to perform some basic operations and process the data to make it ready for our neural network. 
- The dimension of the training data is (60000*28*28).
- One more dimension is needed for the CNN model so we reshape the matrix to shape (60000*28*28*1).


### normalizing the image

```python
x_train=x_train.astype(np.float32)/255
```

```python
x_test=x_test.astype(np.float32)/255
```

###  reshape the dimension

```python
x_train=np.expand_dims(x_train,-1)
x_test=np.expand_dims(x_test,-1)
```

```python
x_train.shape
```

# converting classes to one hot vectors

```python
y_train=keras.utils.to_categorical(y_train)
```

```python
y_test=keras.utils.to_categorical(y_test)
```

# model creation
- Its time for the creation of the CNN model for this Python-based data science project. 
- A convolutional layer and pooling layers are the two wheels of a CNN model. 
- The reason behind the success of CNN for image classification problems is its feasibility with grid structured data. 
- We will use the Adadelta optimizer for the model compilation.

```python
model= Sequential()
```

```python
model.add(Conv2D(32,(3,3),input_shape=(28,28,1),activation='relu'))
model.add(MaxPool2D((2,2)))
```

```python
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPool2D((2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Dense(10,activation="softmax"))
```

```python
#getting model summary

model.summary()
```

```python
# compiling the model
model.compile(optimizer='adam',loss=keras.losses.categorical_crossentropy,metrics=['accuracy'])
```

 ### - reason of using adam as an optimizer
 - we have used "adam as an optimizer" because  As an optimizer  Adam combines the concepts of momentum and RMSprop.
 - It uses adaptive learning rates and maintains an exponentially decaying average of past gradients as well as their squares.

```python
# callback
#A callback is a function that is passed as an argument to other function.
from keras.callbacks import EarlyStopping,ModelCheckpoint
```

```python
#Early stopping is an optimization technique used to reduce overfitting without compromising on model accuracy.
#main purpose of using early stopping in this model is  is to stop training before a model starts to overfit.
es=EarlyStopping(monitor='val_acc',min_delta=0.01,patience=4,verbose=1)
```

```python
#model Check Point-The ModelCheckpoint callback class allows you to define where to checkpoint the model weights,
#how to name the file, and under what circumstances to make a checkpoint of the model.
#The API allows you to specify which metric to monitor, such as loss or accuracy on the training or validation dataset
mc=ModelCheckpoint("./bestmodel.h5",monitor="val_acc",verbose=1,save_best_only=True)
```

```python
cb=[es,mc]
```

# Train the model


- To start the training of the model we can simply call the model.fit() function of Keras. 
- It takes the training data, validation data, epochs, and batch size as the parameter.
- The training of model takes some time.
- After succesful model training, we can save the weights and model definition in the ‘mnist.h5’ file

```python
his=model.fit(x_train,y_train,epochs=5,validation_split=0.3)
```

```python
print("The model has successfully trained")
model.save('mnist.h5')
print("Saving the bot as mnist.h5")
```

# Evaluate the model

### WHAT IS PIL

```python
from PIL  import ImageGrab,Image
```

```python
import numpy as np
```

```python
model = load_model('mnist.h5')

def predict_digit(img):
    #resize image to 28x28 pixels
    img = img.resize((28,28))
    #convert rgb to grayscale
    img = img.convert('L')
    img = np.array(img)
    #reshaping to support our model input and normalizing
    img = img.reshape(1,28,28,1)
    img = img/255.0
    #predicting the class
    res = model.predict([img])[0]
    return np.argmax(res), max(res)

class App(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)

        self.x = self.y = 0

        # Creating elements
        self.canvas = tk.Canvas(self, width=300, height=300, bg = "white", cursor="cross")#creating the diplay canvas
        #creating label inside result area as soon the result pops up it changes to the required value
        self.label = tk.Label(self, text="Thinking..", font=("Helvetica", 48))
        #creating first button to regconize the input and give the result accordingly
        self.classify_btn = tk.Button(self, text = "Recognise", command =         self.classify_handwriting) 
        #creating second button to clear the input
        self.button_clear = tk.Button(self, text = "Clear", command = self.clear_all)

        # Grid structure
        #grid basically is a geometry manager organizes widgets in a table-like structure in the parent widget.
        self.canvas.grid(row=0, column=0, pady=2, sticky=W, )
        self.label.grid(row=0, column=1,pady=2, padx=2)
        self.classify_btn.grid(row=1, column=1, pady=2, padx=2)
        self.button_clear.grid(row=1, column=0, pady=2)
        
        #bind is defined as a Tkinter function for binding events which may occur by initiating the code written in 
        #the program and to handle such events occurring in the program are handled by the binding 
        #function where Python provides a binding function known as bind() 

        #self.canvas.bind("<Motion>", self.start_pos)
        self.canvas.bind("<B1-Motion>", self.draw_lines)

    def clear_all(self):
        self.canvas.delete("all")

    def classify_handwriting(self):
        HWND = self.canvas.winfo_id() # get the handle of the canvas
        rect = win32gui.GetWindowRect(HWND) # get the coordinate of the canvas
        im = ImageGrab.grab(rect)

        digit, acc = predict_digit(im)
        self.label.configure(text= str(digit)+', '+ str(int(acc*100))+'%')

    def draw_lines(self, event):
        self.x = event.x
        self.y = event.y
        r=8
        self.canvas.create_oval(self.x-r, self.y-r, self.x + r, self.y + r, fill='black')

app = App()
mainloop()
```
